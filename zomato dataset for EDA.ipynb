```python
country_code=pd.read_excel('Country-Code.xlsx')
country_code
```

```python
final_df=pd.merge(df,country_code,on='Country Code',how='left')
```

```python
final_df.Country.value_counts()
```

```python
a=final_df.Country.value_counts().index
a
```

```python
b=final_df.Country.value_counts().values
b
```

```python
colors='yellow','green','red'
title="Top Three Transaction"
explode=[0,0.5,0.5]
plt.pie(b[:3],labels=a[:3],autopct='%1.2f%%',colors=colors,explode=explode)
plt.show()
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
Ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Result'})
Ratings
```

```python
plt.bar(Ratings['Aggregate rating'],Ratings['Rating Result'])
plt.xticks=(Ratings['Aggregate rating'])
plt.yticks=(Ratings['Rating Result'])
plt.show(
)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.yticks=(range(1,2500,400))
plt.figure(figsize=(15,15))
palette=['white','red','orange','yellow','green','darkgreen']
sns.barplot(x=Ratings["Aggregate rating"],y=Ratings['Rating Result'],palette=palette,hue=Ratings['Rating color'])
plt.show()
```

```python
palette=['white','red','orange','yellow','green','darkgreen']
sns.countplot(x=Ratings['Rating color'],palette=palette)
```
